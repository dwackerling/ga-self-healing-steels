```python
manual_cases = [vals_min, vals_max, vals_mid, vals_offgrid, vals_outside]

for i, vals in enumerate(manual_cases, start=1):
    clean = be.sanitize_values(vals)
    code = be.encode(vals)
    dec = be.decode(code)
    ok = clean == dec
    print(f"CASE {i}: round-trip exact = {ok}")
    if not ok:
        print("SANITIZED:")
        pprint.pprint(clean)
        print("DECODED:")
        pprint.pprint(dec)
        print()
```

```python
rng = random.Random(123)

n = 1000
all_ok = True
negative_fe = 0

for _ in range(n):
    code = be.random_valid_code(rng)
    dec = be.decode(code)
    recode = be.encode(dec)
    dec2 = be.decode(recode)

    if dec != dec2:
        all_ok = False
        print("Mismatch found:")
        print(dec)
        print(dec2)
        break

    if dec["Fe"] < 0:
        negative_fe += 1

print(f"Round-trip stability OK: {all_ok}")
print(f"Negative Fe cases: {negative_fe} / {n}")
```

```python
rng = random.Random(7)

original = be.random_valid_code(rng)
bits = list(be.to_bitstring(original))

# flip 5 random bits
flip_positions = rng.sample(range(len(bits)), 5)
for pos in flip_positions:
    bits[pos] = "1" if bits[pos] == "0" else "0"

corrupted_bits = "".join(bits)
corrupted_code = be.from_bitstring(corrupted_bits)
sanitized_code = be.sanitize_code(corrupted_code)

print("ORIGINAL CODE:")
print(original)
print("ORIGINAL DECODED:")
pprint.pprint(be.decode(original))
print()

print("CORRUPTED CODE:")
print(corrupted_code)
print("CORRUPTED DECODED:")
pprint.pprint(be.decode(corrupted_code))
print()

print("SANITIZED CODE:")
print(sanitized_code)
print("SANITIZED DECODED:")
pprint.pprint(be.decode(sanitized_code))
```

```python
def run_encoder_smoke_test():
    rng = random.Random(2026)
    n = 1000

    results = {
        "roundtrip_ok": True,
        "negative_fe_count": 0,
        "examples_checked": 0,
    }

    for _ in range(n):
        code = be.random_valid_code(rng)
        dec = be.decode(code)
        recode = be.encode(dec)
        dec2 = be.decode(recode)

        if dec != dec2:
            results["roundtrip_ok"] = False
            print("ERROR: roundtrip mismatch")
            print("dec :", dec)
            print("dec2:", dec2)
            break

        if dec["Fe"] < 0:
            results["negative_fe_count"] += 1

        results["examples_checked"] += 1

    return results

summary = run_encoder_smoke_test()
pprint.pprint(summary)
```
